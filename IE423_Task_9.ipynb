```python
# Use pipelines to access pre-trained models
from transformers import pipeline
```

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters
#prmStatement = 'We are happy to go on vacation this spring break.'
prmStatement1 = ['It is a beautiful day outside']
prmStatement2 = ['I do not enjoy my job']
prmStatement3 = ['I apologize']
prmStatement4 = ['I was upset but now I am okay']


# Predict
print("statement 1:", mdlSnt(prmStatement1))
print("statement 2:", mdlSnt(prmStatement2))
print("statement 3:", mdlSnt(prmStatement3))
print("statement 4:", mdlSnt(prmStatement4))
```

Negative labels seem to be associated with negating or negative words like "not" and "apologize". The model seems to understand that the concluding sentiment is more important than the beginning as seen by the label for statement 4.

```python
# Model
mdlSnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmStatement = 'Ellos no hablan nadie' #spanish for: they don't speak to anyone

# Predict
mdlSnt(prmStatement)
```

```python
# Model
mdlQa = pipeline("question-answering")

# Parameters
f = """The Amazon Rainforest, often referred to as the "lungs of the Earth," spans approximately 5.5 million square kilometers across nine countries in South America. It is the largest rainforest in the world, containing over 390 billion individual trees divided into 16,000 species. The Amazon plays a crucial role in regulating the global climate by absorbing large amounts of carbon dioxide and releasing oxygen. It is home to an estimated 10% of the world's known biodiversity, including thousands of species of plants, insects, birds, and mammals, many of which are not found anywhere else on the planet."""
question1 = "How many species of trees are in the Amazon Rainforest?"
question2 = "What is the approximate size of the Amazon Rainforest?"
question3 = "What role does the Amazon Rainforest play in regulating the global climate?"


# Predict
print("Answer 1:", mdlQa(question = question1, context = f))
print("Answer 2:", mdlQa(question = question2, context = f))
print("Answer 3:", mdlQa(question = question3, context = f))
```

Higher scores for more precise questions with clear answers from the text.
