```python
dfRatings2['UserID'] = dfRatings1.columns[1:].str.extract('(\d+)').astype(int)  # Extract number from column names

# Melt dataframe to long format
dfRatingsLong = dfRatings2.melt(id_vars=['JokeId'], var_name='UserID', value_name='Rating')

# Display the first few rows of the long format dataframe
dfRatingsLong
```

```python
# Drop rows with NaN values in 'Rating' column
dfRatingsLong = dfRatingsLong.dropna(subset=['Rating'])

# Display the first few rows of the cleaned dataframe
dfRatingsLong.shape
```

```python
# Prepare data into Surprise library format

!pip3 install scikit-surprise #or !conda install -c conda-forge scikit-surprise
from surprise import Dataset
from surprise import Reader
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(0,5))
X = Dataset.load_from_df(dfRatingsLong[['UserID', 'JokeId', 'Rating']], reader)
X_train, X_test = train_test_split(X, test_size=.25)
```

```python
# Define SVD model

from surprise import SVD

mdlSvdRatings = SVD()
```

```python
# Fit SVD model

mdlSvdRatings.fit(X_train)
test_pred = mdlSvdRatings.test(X_test)
```

```python
# Evalute SVD accuracy

from surprise import accuracy

accuracy.rmse(test_pred)
```

```python
# Tune hyperparameters

from surprise.model_selection import GridSearchCV

param_grid = {'n_epochs': [5, 10, 15], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(X)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# Cross-validate

from surprise.model_selection import cross_validate

cross_validate(mdlSvdRatings, X, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
